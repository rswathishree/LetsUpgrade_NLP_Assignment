# Removing HTML Tags

```python
import re
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
text= text.apply(lambda x: remove_html_tags(x))
```

```python
print(text[0])
```

# Removing Punctuation

```python
import string
regular_punct = list(string.punctuation)
list1=[':','@','_','=','-']
for element in list1:
    if element in regular_punct:
        regular_punct.remove(element)

def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()
text= text.apply(lambda x:remove_punctuation(x,regular_punct))
print(text[0])
```

# Converting Lowercase

```python
text= text.apply(lambda x: x.lower())
```

```python
print(text[0])
```

# Tokenization

```python
#defining function for tokenization
def tokenization(text):
    return text.split()
text= text.apply(lambda x: tokenization(x))
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
print(text[0])
```

# Stop Word Removal

```python
#importing nlp library
import nltk
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#applying the function
text= text.apply(lambda x:remove_stopwords(x))
```

```python
print(text[0])
```

# Stemming

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text
textsem=text.apply(lambda x: stemming(x))
```

```python
print(textsem[0])
```

# Lemmatization 

```python
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
#defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text
textlem=text.apply(lambda x:lemmatizer(x))
```

```python
print(textlem[0])
```

# forming sentence

```python
cont = []
for row in textlem:
    line = ''
    for word in row:
        line = line + ' ' + word
    cont.append(line)
```

```python
cont[0]
```

# Bag of word

```python
cv = CountVectorizer(max_features=10000)
```

```python
cv.fit_transform(cont)
```
